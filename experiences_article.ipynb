```python
import PyChest
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
```

```python
seq = []
seq.append(np.random.binomial(1, p=0.2, size=2000))
seq.append(np.random.binomial(1, p=0.7, size=4500))
seq.append(np.random.binomial(1, p=0.2, size=1500))
seq = np.concatenate(seq)
```

```python
seq.shape
```

```python
min_distance = 0.125
estimates = PyChest.list_estimator(seq, min_distance)
```

```python
estimates
```

```python
process_count = 2
estimates2 = PyChest.find_changepoints(seq, min_distance, process_count)
```

```python
estimates2
```

```python
N = 25
seq = np.convolve(seq, np.ones(N)/N, mode='valid')
```

```python
estimates3 = PyChest.list_estimator(seq, min_distance)
```

```python
estimates3
```

```python
plt.plot(seq)
plt.axvline(1989, color = 'red', linestyle = 'dashed')
plt.axvline(6489, color = 'red', linestyle = 'dashed')
plt.axvline(4582, color = 'red', linestyle = 'dashed')
plt.show()
```
